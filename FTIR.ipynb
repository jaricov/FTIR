```python
import csv
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
from matplotlib.text import OffsetFrom
from matplotlib import pyplot
import numpy as np

x_TPU=[]
y_TPU=[]
with open('TPU.txt','r') as csvfile:
    file = csv.reader(csvfile, delimiter='\t')
    for row in file:
        x_TPU.append(float(row[0]))
        y_TPU.append(float(row[1]))
        
x_C3=[]
y_C3=[]
with open('C3minusTPU.txt','r') as csvfile:
    file = csv.reader(csvfile, delimiter='\t')
    for row in file:
        x_C3.append(float(row[0]))
        y_C3.append(float(row[1]))
        
x_A2=[]
y_A2=[]
with open('A2minusTPU.txt','r') as csvfile:
    file = csv.reader(csvfile, delimiter='\t')
    for row in file:
        x_A2.append(float(row[0]))
        y_A2.append(float(row[1]))
        
pyplot.ion()
fig=pyplot.figure(figsize=(6,7))
pyplot.subplots_adjust(hspace=0)
fig.text(0.5, 0.07, 'Wavelength [$cm^{-1}$]', ha='center', va='center',fontsize=11)
fig.text(0.1, 0.5, 'Transmitance % [Arb. Unit]', ha='center', va='center', rotation='vertical',fontsize=11)

ax1=fig.add_subplot(3,1,1)
ax2=fig.add_subplot(3,1,2)
ax3=fig.add_subplot(3,1,3)

ax1.plot(x_TPU,y_TPU,label='TPU',c='b')
ax2.plot(x_A2,y_A2,label='GO with TPU reference',c='r')
ax3.plot(x_C3,y_C3,label='EEG with TPU reference',c='g')
ax1.set_xlim(4100, 300)
ax1.set_ylim(110, 30)
ax1.axes.get_yaxis().set_visible(False)
ax2.set_xlim(4100, 300)
ax2.set_ylim(40, -19)
ax2.axes.get_yaxis().set_visible(False)
ax3.set_xlim(4100, 300)
ax3.set_ylim(40, -19)
ax3.axes.get_yaxis().set_visible(False)
ax1.legend(loc='upper left',frameon=False)
ax2.legend(loc='upper left',frameon=False)
ax3.legend(loc='upper left',frameon=False)
fig.savefig("FTIR.png")
plt.show()
```
